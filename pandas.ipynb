### Pandas-Dataframe And Series

Pandas is a powerfull data manipulation library in python,widely used for data analysis and data cleaning. It providdes two primary data structures: Series and DataFrame. A series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes(rows and columns).

```python
import pandas as pd
```

```python
## SERIES 
# A Pandas series is a one-dimensional array like obkject that can hold any data type. It is similar as column in a table

data = [1,2,3,4,5]
series = pd.Series(data)
print("Series: \n",series) # Here default index is: 0,1,2,3,4...
print(type(series))
```

```python
## Create a series form dictionary
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict) # Here indexes are a,b,c
# Keys from dict treated as Index, values are treated as Series
```

```python
# We can also achive the Dictionary like custom indexes without using dictionary
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## Creae A Dataframe from a dictionary of List

data={
    'Name':['Krish','John','Jack'],
    'Age':[25,30,45],
    'City':['Bangalore','New york','Florida']
}

dataFrame = pd.DataFrame(data)
print(dataFrame)
print(type(dataFrame))
```
