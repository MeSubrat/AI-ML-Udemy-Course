### Pandas-Dataframe And Series

Pandas is a powerfull data manipulation library in python,widely used for data analysis and data cleaning. It providdes two primary data structures: Series and DataFrame. A series is a one-dimensional array-like object, while a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure with labeled axes(rows and columns).

```python
## SERIES 
# A Pandas series is a one-dimensional array like obkject that can hold any data type. It is similar as column in a table

data = [1,2,3,4,5]
series = pd.Series(data)
print("Series: \n",series) # Here default index is: 0,1,2,3,4...
print(type(series))
```

```python
## Create a series form dictionary
data = {'a':1,'b':2,'c':3}
series_dict = pd.Series(data)
print(series_dict) # Here indexes are a,b,c
# Keys from dict treated as Index, values are treated as Series
```

```python
# We can also achive the Dictionary like custom indexes without using dictionary
data = [10,20,30]
index = ['a','b','c']
pd.Series(data,index=index)
```

### Data Manipulation With Pandas And Numpy

```python
import pandas as pd

df = pd.read_csv('employess_data.csv')
# df.head(15)
# Handling Missing Values
# df.isnull().any(axis=1)
df.isnull().sum() 
#isnull() returns a DataFrame (or Series) of boolean values where True indicates a missing (null/NaN) value and False indicates a non-null value.
#isnull().sum() : When applied after isnull(), sum() counts the number of missing values (True) in each column (by default).
# #Because:
# True → 1
# False → 0

```

## ffill() vs fillna()

# ffill: 
-  ffill (forward fill) is a method/strategy used to fill missing values using the previous valid value.
-  It is not a standalone function.
-  It is used inside fillna() (or via .ffill() shortcut).

# fillna: 
- fillna() is a pandas function used to replace missing values using a value or a method (like ffill, bfill).

```python
## Filling missing values with 0
df_filled = df.fillna(0)
df_filled
```

```python
## filling missing values with the mean of the column 
df['Age_fillNa'] = df['Age'].fillna(df['Age'].mean())
## First we go to the Age column, then in that column we fill the NaN values present in that column with mean values of Age column
df
```

```python
# Renaming columns
df = df.rename(columns={'Name':'First Name'})
df.head()
```

```python
## Change datatypes
df['Salary_New'] = df['Salary'].fillna(df['Salary'].mean()).astype(int) 
# Here I have used the fillna cause there is NaN values(null values) present in the Salary column.
df.head()
```

```python
df['New Age'] = df['Age'].apply(lambda x : x*2)
# By this .apply() we can actually use any functions inside this .apply().
df
```

## Important Point for grouping
- NaN salaries are automatically ignored in mean()
- If a city has all NaN salaries, result will be NaN
- Rows where City is NaN are excluded from grouping

```python
## Data aggregating and grouping
grouped_mean = df.groupby('City')['Salary'].mean()
print(grouped_mean)
```

```python
# We can group by multiple values by passing them as a list
df.groupby(['City','Department'])['Salary'].mean()
```

```python
## Multiple aggegate function
grouped_agg = df.groupby('Department')['Salary'].agg(['mean','sum','count'])
grouped_agg
```

```python
#Merging and joining dataFrames
# Create sample dataFrames

df1 = pd.DataFrame({'key': ['A','B','C'],'value1':[1,2,3]})
df2 = pd.DataFrame({'key': ['A','B','D'],'value2':[4,5,6]})
```

```python
## Merge Dataframes on the 'Key Column'
pd.merge(df1,df2,on='key',how='inner')
```

```python
pd.merge(df1,df2,on='key',how='outer')
```

```python
pd.merge(df1,df2,on='key',how='left')
```

```python
pd.merge(df1,df2,on='key',how='right')
```
